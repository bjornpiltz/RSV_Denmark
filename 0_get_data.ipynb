# Get RSV data from SSI
https://experience.arcgis.com/experience/220fef27d07d438889d651cc2e00076c/page/RS-virus/

```python
import requests, urllib, json, pandas as pd

age_groups = ["01. 0-5 måneder", "02. 6-11 måneder",  "03. 1 år",  
              "04. 2 år",        "05. 3-5 år",        "06. 6-14 år",
              "07. 15-44 år",    "08. 45-64 år",      "09. 65-74 år",
              "10. 75-84 år",    "11. 85+ år"]

df = []

for age_grp in age_groups:
    print(age_grp)
    for value_type in ["cases", "tested", "admissions", "deaths", "incidens"]:
        q = urllib.parse.quote("(agegrp=\'" + age_grp + "\')", safe = "()")
        s = ("https://services5.arcgis.com/Hx7l9qUpAnKPyvNz/arcgis/rest/services/DB_rsvirus_data_gdb/FeatureServer/2/query?f=json&cacheHint=true&groupByFieldsForStatistics=week_sort%2CSeason&orderByFields=week_sort%20ASC&outFields=*&outStatistics=%5B%7B%22onStatisticField%22%3A%22" +
             value_type + 
             "%22%2C%22outStatisticFieldName%22%3A%22value%22%2C%22statisticType%22%3A%22sum%22%7D%5D&resultType=standard&returnGeometry=false&spatialRel=esriSpatialRelIntersects&where=(region%3D%27Alle%27)%20AND%20" +
             q + "%20AND%20(gender%3D%27Alle%27)")

        content = content = requests.get(s).json()
        for item in content["features"]:
            d = {
                'week_sort' : item['attributes']['week_sort'],
                'value' : item['attributes']['value'],
                'type' : value_type,
                'Season' : item['attributes']['Season'],
                'Age Group' : age_grp,
            }
            df.append(d)
df = pd.DataFrame(df).set_index("week_sort")

df.to_csv("data/rsv_denmark.csv")

df
```
